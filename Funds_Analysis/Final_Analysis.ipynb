```python
# Supress Warnings
import warnings
warnings.filterwarnings('ignore')

# import pandas, pd is an alias
# import numpy, np is an alias
import pandas as pd
import numpy as np
```

```python
#importing companies file
companies= pd.read_csv(r'C:\Users\Admin\Desktop\Upgrad files\CASE_STUDY_1\Downloads\companies.csv', encoding = 'latin')
companies = companies.applymap(lambda x: str(x).encode('utf-8').decode('ascii', 'ignore'))
companies
```

```python
#importing rounds2 file
rounds2 = pd.read_csv(r'C:\Users\Admin\Desktop\Upgrad files\CASE_STUDY_1\Downloads\rounds2.csv', encoding = 'latin')
rounds2 = rounds2.applymap(lambda x: str(x).encode('utf-8').decode('ascii', 'ignore'))
rounds2
```

```python
# unique companies in rounds2
rounds2['company_permalink']=rounds2['company_permalink'].str.lower()
len(rounds2['company_permalink'].unique())
```

```python
# unique companies in companies
companies['permalink']= companies['permalink'].str.lower()
len(companies['permalink'].unique())
```

```python
#creating master_frame
master_frame = rounds2.join(companies.set_index('permalink'), on='company_permalink')
print(len(master_frame))
master_frame
```

```python
# master_frame info to check NULL fields
print(master_frame.shape)
print(master_frame.info())
```

```python
# Dropping NaN from 'raised_amount_usd' column
master_frame = master_frame.dropna(subset=['raised_amount_usd'])
master_frame
```

```python
# converting 'raised_amount_usd' to amounts in million
master_frame['country_code'] = master_frame['country_code'].replace("nan", np.NaN, regex=True)
master_frame = master_frame.dropna(subset=['raised_amount_usd'])
master_frame['raised_amount_usd']=master_frame['raised_amount_usd'].astype(float)/1000000
master_frame.groupby('funding_round_type')['raised_amount_usd'].mean()
```

```python
#filter master frame with venture type
venture = master_frame.loc[master_frame['funding_round_type'] =='venture']
venture = venture.dropna(subset=['country_code'])
venture
```

```python
#top 9 countries in descending order
top9 = venture.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending = False).head(9)
top9
```

```python
#extracting primary category
venture['primary_sector']= venture['category_list'].str.split('|').str[0]
venture
```

```python
mapping= pd.read_csv('mapping.csv', encoding = 'latin')
mapping
```

```python
mapping['category_list'] = mapping['category_list'].str.replace('0', 'na')
mapping
```

```python
#mapping main sector
mapping['main_sector']=np.nan
mapping.loc[mapping['Automotive & Sports']==1, 'main_sector']= 'Automotive & Sports'
mapping.loc[mapping['Cleantech / Semiconductors']==1, 'main_sector']= 'Cleantech / Semiconductors'
mapping.loc[mapping['Entertainment']==1, 'main_sector']= 'Entertainment'
mapping.loc[mapping['Health']==1, 'main_sector']= 'Health'
mapping.loc[mapping['Manufacturing']==1, 'main_sector']= 'Manufacturing'
mapping.loc[mapping['News, Search and Messaging']==1, 'main_sector']= 'News, Search and Messaging'
mapping.loc[mapping['Others']==1, 'main_sector']= 'Others'
mapping.loc[mapping['Social, Finance, Analytics, Advertising']==1, 'main_sector']= 'Social, Finance, Analytics, Advertising'

mapping_sector= mapping[['category_list','main_sector']]
mapping_sector
```

```python
#primary sector and main sector merged together
venture= pd.merge(venture,mapping_sector, left_on='primary_sector', right_on='category_list', how='inner')
venture
```

```python
#creating data frames for 3 countries based on country and funding amount 5-15M
#USA
D1= pd.DataFrame(venture.loc[(venture['country_code']=='USA') & ((venture['raised_amount_usd'] >= 5) & (venture['raised_amount_usd'] <= 15))])
D1
```

```python
#total amount of investment
temp1= D1.groupby('main_sector', as_index=False).sum()
temp1= temp1.rename(columns={"raised_amount_usd":"Total Investments"})
temp1
D1= D1.join(temp1.set_index(['main_sector']), on=['main_sector'], how= 'left')
D1
```

```python
#count of investments

count=D1.groupby('main_sector', as_index=False)['raised_amount_usd'].count()
count= count.rename(columns={"raised_amount_usd":"Total Count"})
count

#final df for USA
D1= D1.join(count.set_index(['main_sector']), on=['main_sector'], how= 'left')
D1
```

```python
#total number of investments
print (D1['raised_amount_usd'].count())
#total amount of investment
print (D1['raised_amount_usd'].sum())
#top 3 sectors based on count
count.sort_values(['Total Count'], ascending=False)
```

```python
#finding top company for sector 1
D1[D1['main_sector']=="Others"].groupby('name')['raised_amount_usd'].sum().sort_values(ascending=False)
```

```python
#finding top comapny for sector 2
D1[D1['main_sector']=="Social, Finance, Analytics, Advertising"].groupby('name')['raised_amount_usd'].sum().sort_values(ascending=False)
```
